```python
%pylab inline
import csv
```

```python
with open ('Food_Carbon_Emission.csv', mode='r') as file1:
    reader=csv.reader(file1)
    food_carbon=list(reader)

food_carbon[1:]
```

```python
def get_carbon(food_present):
    result=[]
    
    for item in food_present:
        results=[]
        for line in food_carbon[1:]:
            if line[0]==item:
                co2_per_serving=float(line[5])
                co2_per_gram=float(line[5])/float(line[2])
                results=results+[item,co2_per_serving,co2_per_gram]
                
                result=result+[results]
    return result
```

```python
print(get_carbon(['BANANAS','AVOCADOS']))
```

```python
def carb_sum(get_carbon_output):
    sum_c=0
    
    for i in get_carbon_output:
        sum_c=sum_c+ i[1]
    return sum_c
```

```python
foods = input("foods present seperated by spaces: ")
foodlist = foods.split(" ")
carbon_footprint = carb_sum(get_carbon(foodlist))
carbon_footprint
```

```python
def get_water_meat(food):
    'Gets the water footprint'
    result=[]
    water_footprint=0
    for item in food:
        results=[]
        food_item=item[0]
        country=item[1]
        for i in meat_water[0]:
            if i == country:
                food_country=country
        for j in meat_water:
            if j[0]==food_item:
                
                water_footprint=meat_water[meat_water.index(j)][meat_water[0].index(country)]

        results=results+[food_item,water_footprint]
        result=result+[results]
        
    return result
            
```

```python
print(get_water_meat([['BEEF','Austria'],['EGGS','China']]))
```

```python
def meatwater_sum(get_water_meat_output):
    sum_e=0
    
    for i in get_water_meat_output:
        sum_e=sum_e + float(i[1])
    return sum_e
```
